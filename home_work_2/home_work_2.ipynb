# Task about lakes

- Generate 20 random variable from 1 to 100
- Draw the plot of the sequence
- Lets assume each point represent the height and so all plotting is 2d mountains.
- Then consider the unlimited rain from above - cavities become lakes full of water.
- Determine the deepest lake

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2023)
a = np.random.randint(1, 100, 20)

def get_lake_depth(a):
    min = np.argmin(a)
    left = np.argmax(a[:min])
    right = np.argmax(a[min:]) + min
    return left, right

def show_lake(a, left, right):
    water_point = min(a[left], a[right])
    plt.plot(a)
    plt.fill_between(np.arange(left, right+1), water_point, alpha=0.4, color='blue')
    plt.fill_between(np.arange(left, right+1), 0, water_point, alpha=0.5, color='blue', where=(a[left:right+1] == water_point))
    plt.show()

left, right = get_lake_depth(a)

show_lake(a, left, right)
```

```python
# another seed
np.random.seed(530)
a = np.random.randint(1, 100, 20)
left, right = get_lake_depth(a)
show_lake(a, left, right)
```
