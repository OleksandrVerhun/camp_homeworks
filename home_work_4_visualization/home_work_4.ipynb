*Data Cleaning*

```python
import pandas as pd
import numpy as np

file_path = "hw4_files/Car_Insurance_Claim.csv"
df = pd.read_csv(file_path)

# Delete unnecessary indexes
df = df.drop(df.columns[0], axis=1)

# Change "none" to NaN or "Not set"
columns_to_nan = ["ID", "AGE", "GENDER", "RACE", "INCOME", "CREDIT_SCORE", "VEHICLE_YEAR", "POSTAL_CODE", "VEHICLE_TYPE"]
df[columns_to_nan] = df[columns_to_nan].replace("none", np.NaN)
# Change "none" to 0
columns_to_zero = ["VEHICLE_OWNERSHIP", "MARRIED", "CHILDREN", "ANNUAL_MILEAGE", "SPEEDING_VIOLATIONS", "DUIS", "PAST_ACCIDENTS", "OUTCOME"]
df[columns_to_zero] = df[columns_to_zero].replace("none", 0)

# Change to numeric data type
numeric_columns = ["CREDIT_SCORE", "VEHICLE_OWNERSHIP", "MARRIED", "CHILDREN", "ANNUAL_MILEAGE", "SPEEDING_VIOLATIONS", "DUIS", "PAST_ACCIDENTS", "OUTCOME"]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.head(10)
```

*Descriptive Statistics*

```python
df.describe()
```

*Data Visualization*

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 3))
sns.histplot(data=df, x='AGE', bins=30, color='lightblue', kde=True)
plt.title('Distribution by age')
plt.xlabel('Age')
plt.ylabel('Number of people')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
g = sns.FacetGrid(df, col='INCOME', col_wrap=4, height=4, margin_titles=True)
g.map(sns.histplot, 'AGE', bins=30, color='skyblue', kde=True)
g.set_axis_labels('Age', 'Number of people')
g.set_titles(col_template='{col_name}')
plt.show()
```

```python
# Delete "none"
df = df[df['DRIVING_EXPERIENCE'] != 'none']
plt.figure(figsize=(5, 5))
g = sns.FacetGrid(df, col='GENDER', col_wrap=4, height=4, margin_titles=True)
g.map(sns.histplot, 'DRIVING_EXPERIENCE', bins=30, color='green', kde=True)
g.set_axis_labels('DRIVING EXPERIENCE', 'Number')
g.set_titles(col_template='{col_name}')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
g = sns.FacetGrid(df, col='MARRIED', col_wrap=4, height=4, margin_titles=True)
g.map(sns.histplot, 'AGE', bins=30, color='pink', kde=True)
g.set_axis_labels('Age', 'Number')
g.set_titles(col_template='{col_name}')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.histplot(data=df, x='ANNUAL_MILEAGE', bins=30, color='red', kde=True)
plt.xlabel('ANNUAL MILEAGE')
plt.ylabel('Number of people')
plt.show()
```

```python
plt.figure(figsize=(5, 5))
g = sns.FacetGrid(df, col='GENDER', col_wrap=4, height=4, margin_titles=True)
g.map(sns.histplot, 'SPEEDING_VIOLATIONS', bins=30, color='gray', kde=True)
g.set_axis_labels('SPEEDING VIOLATIONS', 'Number')
g.set_titles(col_template='{col_name}')
plt.show()
```

### Conclusions

- **AGE**: Most people are aged 26-39.
- **INCOME**: In the upper class, the majority of people are aged 40 and older. In the middle and working class, most people are aged 26-39. And in poverty, the majority of people are aged 16-25.
- **DRIVING EXPERIENCE**: Both men and women have the same driving experience. And most people have 0-9 years of experience.
- **MARRIED**: More married people at an older age.
- **ANNUAL MILEAGE**: The most common value is 12000.
